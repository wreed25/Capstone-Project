## Data Cleaning

Before we begin initial data exploration, we must make sure we have clean data to test and most importantly, we need to join our player stats data with our contract data so we can understand the full scope of the task at hand.



## Player Stats Cleaning

We will need to make sure there are no empty rows, missing GSIS IDs, check missing or invalid values, as well as gte a subset of Wide Receiver Stats to prepare for joining the data. The most notable of these is the missing GSIS ID or else we will be missing a full players stats or even worse, won't be able to join the data successfully

```python
# Confirm that the existing objects are available

print("Player statistics:", player_stats.shape)
print("Player metadata:", players.shape)
print("WR contracts:", wr_contracts.shape)
```

```python
# Confirm that the player_id and gsis_id columns are of the same type

print("player_stats.player_id type:")
print(player_stats.schema["player_id"])

print("\nplayers.gsis_id type:")
print(players.schema["gsis_id"])

print("\nwr_contracts.gsis_id type:")
print(wr_contracts.schema["gsis_id"])
```

```python
# Create the subset for WR player statistics

wr_player_stats = player_stats.filter(
    pl.col("position") == "WR"
)

print("WR player-stat rows:", wr_player_stats.height)
```

```python
# Select the columns of interest for WR player statistics

wr_player_stats = wr_player_stats.select([
    "player_id",
    "player_name",
    "player_display_name",
    "position",
    "season",
    "season_type",
    "recent_team",
    "games",
    "receptions",
    "targets",
    "receiving_yards",
    "receiving_tds",
    "receiving_fumbles",
    "receiving_fumbles_lost",
    "receiving_air_yards",
    "receiving_yards_after_catch",
    "receiving_first_downs",
    "receiving_epa",
    "receiving_2pt_conversions",
    "receiving_10",
    "receiving_16",
    "receiving_20",
    "receiving_40",
    "racr",
    "target_share",
    "air_yards_share",
    "wopr",
    "fantasy_points",
    "fantasy_points_ppr"
])
```

```python
# Renaming player_id to gsis_id to match the other dataframes for merging

wr_player_stats = wr_player_stats.rename({
    "player_id": "gsis_id"
})
```

```python
# Remove rows with null or empty gsis_id values to ensure data integrity

wr_player_stats = wr_player_stats.filter(
    pl.col("gsis_id").is_not_null() &
    (pl.col("gsis_id").str.strip_chars() != "")
)
```

```python
# Standardizing the gsis_id column by stripping any leading or trailing whitespace characters

wr_player_stats = wr_player_stats.with_columns(
    pl.col("gsis_id")
      .str.strip_chars()
      .alias("gsis_id")
)
```

```python
# Checking for duplicate player-season combinations in the WR player statistics

player_stats_duplicates = (
    wr_player_stats
    .group_by(["gsis_id", "season"])
    .len()
    .filter(pl.col("len") > 1)
)

print("Duplicate player-season combinations:")
print(player_stats_duplicates.shape)

player_stats_duplicates.head()
```

```python
# Checking for any missing values in the WR player statistics dataframe

player_stats_missing = wr_player_stats.select([
    pl.col(column).null_count().alias(column)
    for column in wr_player_stats.columns
]).transpose(
    include_header=True,
    header_name="column",
    column_names=["missing_count"]
).sort(
    "missing_count",
    descending=True
)

player_stats_missing
```

## Player Contracts Cleaning

We will follow a very similar process as we did for the Player Stats Cleanup since we plan on joining the data we must make sure everything aligns well

```python
# Creating a subset of the WR contracts dataframe with only the columns of interest for analysis

wr_contracts_clean = wr_contracts.select([
    "player",
    "position",
    "team",
    "is_active",
    "year_signed",
    "years",
    "value",
    "apy",
    "guaranteed",
    "apy_cap_pct",
    "inflated_value",
    "inflated_apy",
    "inflated_guaranteed",
    "gsis_id",
    "college",
    "draft_year",
    "draft_team"
])
```

```python
# Cleaning the gsis_id column in the WR contracts dataframe by stripping any leading or trailing whitespace characters

wr_contracts_clean = wr_contracts_clean.filter(
    pl.col("gsis_id").is_not_null() &
    (pl.col("gsis_id").str.strip_chars() != "")
)
```

```python
wr_contracts_clean = wr_contracts_clean.with_columns(
    pl.col("gsis_id")
      .str.strip_chars()
      .alias("gsis_id")
)
```

```python
# Checking for any missing or blank GSIS IDs in the WR contracts dataframe

print("Contract rows after GSIS ID cleaning:",
      wr_contracts_clean.height)

print(
    "Missing GSIS IDs:",
    wr_contracts_clean
    .select(pl.col("gsis_id").is_null().sum())
    .item()
)

print(
    "Blank GSIS IDs:",
    wr_contracts_clean
    .filter(pl.col("gsis_id") == "")
    .height
)
```

```python
# Checking for any missing values in the WR contracts dataframe

contract_missing = wr_contracts_clean.select([
    pl.col(column).null_count().alias(column)
    for column in wr_contracts_clean.columns
]).transpose(
    include_header=True,
    header_name="column",
    column_names=["missing_count"]
).sort(
    "missing_count",
    descending=True
)

contract_missing
```

## Joining the Stats and Contracts data

Now that both dataframes were cleaned and we have a working subset for each. We need to be able to join the two together so we can run exploratory analysis among the whole library of data

```python
# Selecting relevant columns from the players dataframe to create a bridge for merging with player statistics and contracts data

player_id_bridge = players.select([
    "gsis_id",
    "display_name",
    "position",
    "position_group",
    "birth_date",
    "rookie_season",
    "last_season",
    "years_of_experience",
    "latest_team",
    "nfl_id",
    "pfr_id",
    "otc_id",
    "espn_id"
])
```

```python
# Cleaning the gsis_id column in the player_id_bridge dataframe by stripping any leading or trailing whitespace characters and filtering out null or empty values

player_id_bridge = player_id_bridge.with_columns(
    pl.col("gsis_id")
      .str.strip_chars()
      .alias("gsis_id")
).filter(
    pl.col("gsis_id").is_not_null() &
    (pl.col("gsis_id") != "")
)
```

```python
# Checking whether the bridge gsis_id column has any duplicate values, which could indicate potential issues with the player metadata

duplicate_bridge_ids = (
    player_id_bridge
    .group_by("gsis_id")
    .len()
    .filter(pl.col("len") > 1)
)

print("Duplicated GSIS IDs in player metadata:",
      duplicate_bridge_ids.height)
```

```python
# Joining the WR player statistics dataframe with the player metadata bridge on the gsis_id column to enrich the player statistics with additional player information

wr_stats_with_players = wr_player_stats.join(
    player_id_bridge,
    on="gsis_id",
    how="left",
    validate="m:1"
)
```

```python
# Check whether all of the data was matched to players in the bridge by counting the number of rows with null display_name values

stats_identifier_validation = wr_stats_with_players.select([
    pl.len().alias("wr_stat_rows"),
    pl.col("display_name").is_not_null().sum().alias("matched_to_players"),
    pl.col("display_name").is_null().sum().alias("not_matched_to_players")
])

stats_identifier_validation
```

```python
# Joining the dataframes of WR contracts and WR player statistics (with player metadata) on the gsis_id column to create a combined dataset for analysis. The join is performed as a left join to retain all records from the WR contracts dataframe, while adding relevant statistics from the WR player statistics dataframe.

wr_contract_stat_history = (
    wr_contracts_clean
    .join(
        wr_stats_with_players,
        on="gsis_id",
        how="left",
        suffix="_stats"
    )
    .filter(
        pl.col("season") < pl.col("year_signed")
    )
)
```

```python
# Retaining the most recent season before signing a contract for each player by sorting the combined dataframe and grouping by relevant columns, then selecting the last row in each group
# Taking a look at a players stats previous to signing a new contract will show the best connection of a players stats to how they are compensated

wr_contract_prior_season = (
    wr_contract_stat_history
    .sort([
        "gsis_id",
        "year_signed",
        "season"
    ])
    .group_by([
        "gsis_id",
        "year_signed",
        "team",
        "years",
        "value",
        "apy"
    ])
    .tail(1)
)
```

```python
# Checking to see that the join was successful and that the most recent season prior to signing a contract was retained for each player by selecting relevant columns and sorting by year_signed in descending order, then displaying the top 20 rows

wr_contract_prior_season.select([
    "player",
    "gsis_id",
    "year_signed",
    "season",
    "receptions",
    "targets",
    "receiving_yards",
    "receiving_tds",
    "value",
    "apy",
    "guaranteed"
]).sort(
    "year_signed",
    descending=True
).head(20)
```
